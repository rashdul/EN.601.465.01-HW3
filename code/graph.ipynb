```python
import matplotlib.pyplot as plt
```

```python
x = []
y = []
with open("results.txt", 'r') as file:
  for line in file:
    splitted = line.split()
    x.append(int(splitted[0]))
    y.append(float(splitted[1])) ## maybe you want to devide it by length
print(y)
```

```python
plt.scatter(x, y)
plt.xlabel("length")
plt.ylabel("log-prob")
plt.title("performance vs length (times8)")
plt.show()
```
